```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_csv('ex_off.csv',sep=',',skiprows=49953) 
time = np.array(df.iloc[:,0])
voltage = np.array(df.iloc[:,1])

plt.plot(time, voltage, 'o-')
plt.xlim([0,0.02])
plt.show()

dt=np.unique(np.diff(time))[0]
print( np.unique(np.diff(time)))
omegas = np.linspace(1, 100, 100, dtype='int')

amps = []
for iom, om in enumerate(omegas):
    sumr = 0 ; sumi = 0;
    for itmp, it in enumerate(time):
        sumr = sumr + voltage[itmp]*np.cos(2.*np.pi*om*it)*dt
        sumi = sumi + voltage[itmp]*np.sin(2.*np.pi*om*it)*dt
    amps.append(np.sqrt(sumr**2 + sumi**2))



```

```python
plt.plot(omegas, amps, 'o-')
# plt.xlim([1000, 20000])
plt.show()
```

```python
max_index = np.argmax(amps)
print(omegas[max_index])
```

```python
df = pd.read_csv('2024-10-04/100/2000.csv',sep=',',skiprows=49955) 
time = np.array(df.iloc[:,0])
voltage = np.array(df.iloc[:,1])

plt.plot(time, voltage, 'o-')
plt.show()


omegas = np.linspace(100, 4000, 300, dtype='int')

tmax = len(time)
dt=np.unique(np.diff(time))[0]
print( np.unique(np.diff(time)))

amps = []
for iom, om in enumerate(omegas):
    sumr = 0 ; sumi = 0;
    for it in range(tmax):
        time = it*dt
        sumr = sumr + voltage[it]*np.cos(2.*np.pi*om*time)*dt
        sumi = sumi + voltage[it]*np.sin(2.*np.pi*om*time)*dt
    amps.append(np.sqrt(sumr**2 + sumi**2))
```

```python
plt.plot(omegas, amps, 'o-')
# plt.xlim([1000, 20000])
plt.show()
```

```python
max_index = np.argmax(amps)
print(omegas[max_index])
```
